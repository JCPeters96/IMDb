```python
def extract_movie_data(page_urls):
    movie_list = []
    
    for page_url in page_urls:
        res = requests.get(page_url)
        soup= BeautifulSoup(res.text, "html.parser")
        movies = soup.find_all(class_= "lister-item mode-advanced")
        
        for movie in movies:
            def movie_url():
                for url in movies:
                    try:
                        movie_url = "https://imdb.com" + movie.find("a").attrs["href"]
                        return(movie_url)
                    except:
                        continue 
            def movie_title():
                for title in movies:
                    try:
                        movie_title = movie.find('h3').find('a').text
                        return(movie_title)
                    except:
                        continue 
            
            def movie_date():
                for date in movies:
                    try:
                        movie_date = movie.find(class_="lister-item-year text-muted unbold").text
                        movie_date = movie_date.replace('(','').replace(')','')
                        return(movie_date)
                    except:
                        continue             
            #movie_meta = movie.find('span', class_ = 'metascore favorable')
            def movie_rating():
                for rating in movies:
                    try:
                        movie_rating = movie.find('strong').text
                        return(movie_rating)
                    except:
                        continue             
            def movie_genre():
                for genre in movies:
                    try:
                        movie_genre = movie.find(class_="genre").text
                        movie_genre = movie_genre.replace('\n','').replace("  ", "")
                        return(movie_genre)
                    except:
                        continue                                   
            def number_votes():
                for votes in movies:
                    try:
                        num_votes = movie.find_all('span', attrs={'name': 'nv'})
                        number_votes = num_votes[0].text 
                        return(number_votes)
                    except:
                        continue                        
            def movie_grosses():
                for gross in movies:
                    try:
                        movie_grosses = num_votes[1].text if len(num_votes) > 1 else 'N/A' 
                        return(movie_grosses)
                    except:
                        continue                        
            def runtime():
                for time in movies:
                    try:
                        movie_runtime = movie.find(class_="runtime").text 
                        return(movie_runtime)
                    except:
                        continue
                        
            def director():
                for director in movies:
                    try:
                        director = movie.find('p',class_= '').find_all('a')[0].text 
                        return(director)
                    except:
                        continue                      
            def actor_1():
                for actor in movies:
                    try:
                        actor_1 = movie.find('p', class_ = '').find_all('a')[1].text 
                        return(actor_1)
                    except:
                        continue                        
            def actor_2():
                for actor in movies:
                    try:
                        actor_2 = movie.find('p', class_ = '').find_all('a')[2].text 
                        return(actor_2)
                    except:
                        continue                        
            def actor_3():
                for actor in movies:
                    try:
                        actor_3 = movie.find('p', class_ = '').find_all('a')[3].text 
                        return(actor_3)
                    except:
                        continue                        
            def actor_4():
                for actor in movies:
                    try:
                        actor_4 = movie.find('p', class_ = '').find_all('a')[4].text 
                        return(actor_4)
                    except:
                        continue
                        
            #summary = movie.find('p').find(class_="text-muted")
            
                    
            movie_list.append({"movie url": movie_url(),
                              "title" : movie_title(),
                              "date" : movie_date(),
                              "IMDb rating" : movie_rating(),
                              #"metascore" : movie_meta,
                              "genre" : movie_genre(),
                              "director" : director(),
                              "actor_1": actor_1(),
                              "actor_2" : actor_2(),
                              "actor_3" : actor_3(),
                              "actor_4" : actor_4(),
                              "runtime" : runtime(),
                              "votes" : number_votes(),
                              "gross" : movie_grosses()})
                              #"summary" : summary})
            
        #sleep (1)
            
    return movie_list
```
